```python
#removing duplicates and unnamed column 
if 'Unnamed:0' in datav2.columns:
    datav2.drop("Unnamed",axis=1,inplace=True)
#removing true beahivourial duplicates
initial_rows = datav2.shape[0]
subset = [c for c in datav2.columns if c not in ['Time', 'Class']]
datav2.drop_duplicates(subset=subset, inplace=True)
print(f"🧹 Removed {initial_rows - datav2.shape[0]} true duplicate transactions.")

print(f"✅ Data Cleaned. Current Shape: {datav2.shape}")
```

behaviourial duplication and why remvoing time column eliminates it: This allowed the model to find the 8,063 rows where a fraudster (or a system glitch) tried the exact same transaction multiple times at different seconds.

subset = [c for c in datav2.columns if c not in ['Time', 'Class']], here second c is a temporary name for new column where all the columns from datav2 except time and class will be stored , the first c is used for looping through the columns and temporary naming each column that passes the condition as c

```python
#feature engineering
#Transforming raw columns into signals that XGBoost can interpret more effectively.

print("feature 1:: HOUR OF DAY.." )
#converting Time column to hour of day 
#converting seconds into hours
datav2['Hour_of_day']=(datav2 ['Time']//3600) %24 #this is linear hour ,while it does help it does not tell the model about the cyclic nature of hours


#creating is night feature to flag transactions happening between 6 am to 11pm
print("feature 2:: RISKY TIME FLAG...")
datav2['Is_night']=datav2['Hour_of_day'].apply(lambda x:1 if x<=6 or x>=23  else 0) #this creates and checks time window of whether the fraud occured between 11pm to 6 am
print("feature 3:: LOG AMOUNT...")
# converting Amount to log Amount 
datav2['log_amount']=np.log1p(datav2['Amount'])   #it converts transaction amount to logs to minimize the difference between them say 12000rs and 40 rs so absically we squash the values to shorten them
#dropping original Time column to prevent confusion
datav2.drop('Time',axis=1,inplace=True)


```

The Strategy: We need Amount for Financial ROI, but we don't want it for Feature Training.
Block 2 (Feature Engineering): We create Log_Amount. We keep the original Amount here because the X and y variables are still being defined.
The "Photocopy" Step: test_amounts = X_test['Amount'].values. We save the "Real Dollars" into a separate variable. This is critical because you cannot report success in "Log-Dollars." If you catch a thief, you want to tell your boss you saved $5,000, not 8.5 Log-units.

```python
#separating features and target
x=datav2.drop('Class',axis=1)
y=datav2["Class"]
```

```python
#splitting and scaling
#using stratified split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
#saving original amount from test set before scaling it for future calculations
amount_test=x_test['Amount'].values
#removing previous amount column
x_train.drop('Amount',axis=1,inplace=True)
x_test.drop('Amount',axis=1,inplace=True)
```

```python
#scaling using standard scaler
scaler=StandardScaler()
x_train_scaled=pd.DataFrame(scaler.fit_transform(x_train),columns=x_train.columns)
x_test_scaled=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns)
print("data scaled successfully")
```

```python
#applying and training xgboost 
#checking the Class distribution of the dataset
fraud_weight=len(y_train[y_train==0]) / len(y_train[y_train==1])#this is used further by the model to understand about the class imbalance and treat every fraud with more caution
print(f"training complete ,(fraud weight:{fraud_weight:.2f})..")
#training the v2 model
v2_baseline_model=xgb.XGBClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=5,
    learning_rate=0.05,#xgboost builds upon previous tree and tries to fix the mistakes of previous tree, learning rate tells how much the current tree trusts the previous tree also called as shrinkage
    scale_pos_weight=fraud_weight,
    n_jobs=-1,
    eval_metric='aucpr'

)
v2_baseline_model.fit(x_train_scaled,y_train)
print("v2 model trained")
```

learning rate: xgboost builds upon previous trees, the first tree is generally more conservative and only provides low level risks say 10 percent, the second tree takes the input of first tree and applies learning rate combined with its own result it tries to fix the error or imprve the residual (actual value - predicted value), new prediction= old prediction+(learning rate * new predicition)means the 2nd tree will only consider say 0.1 percent of previous tree's result, this prevents overfitting and prevents biasness of one tree if previous one was biased

scale_pos_weight: it tells the model to focus more on the provided attribute



```python
#predicitons and results
test_pred=v2_baseline_model.predict(x_test_scaled)
test_prob=v2_baseline_model.predict_proba(x_test_scaled)[:, 1]
#confusion matrix
true_neg,false_pos,false_neg,true_pos=confusion_matrix(y_test , test_pred).ravel()

#calculating the success metric ( AUPRC)
aup_score= average_precision_score(y_test, test_prob)#the average_precision_score is a function in scikit learning that calculates the auprc at every point ranging between 0 to 1,the closer the score to 1 the better the results
```

```python
#amount saved from fraud
amount_saved=amount_test[(y_test==1) & (test_pred==1)].sum()
false_alarm_cost=false_pos * 5 # $5 for every innocent customer blocked 
net_roi=amount_saved - false_alarm_cost
```

```python
#FINAL BREAKDOWN OF BASELINE MODEL
print(f"SAVED COST ${amount_saved:,.2f}")
print(f"CUSTOMER ANNOYANCE COST${false_alarm_cost:,.2f}( {false_pos} false alarms)")
print(f"AUPRC SCORE {aup_score:,.4f}")
print(f"NET ROI ${net_roi:,.2f}")
```
